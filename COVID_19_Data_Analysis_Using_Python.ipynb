```python
data = data[['continent', 'location', 'date', 'total_cases', 'total_deaths', 'gdp_per_capita', 'human_development_index']]
print(data)
```

```python
data.drop_duplicates(inplace=True)                # a. Remove duplicates
print(data.isnull().sum())                        # b. Check missing values
data.dropna(subset=['continent'], inplace=True)  # c. Drop where continent is missing
data.fillna(0, inplace=True)                      # d. Fill other missing values with 0
```

```python
data['date'] = pd.to_datetime(data['date'], format="%Y-%m-%d")  # a. Convert to datetime
data['month'] = data['date'].dt.month              # b. Extract month
print(data['date'].head())
print(data['month'].head())

```

```python
data_groupby = data.groupby('continent').max().reset_index()
```

```python
data_groupby['total_deaths_to_total_cases'] = data_groupby['total_deaths'] / data_groupby['total_cases']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# a. Univariate analysis
sns.histplot(data['gdp_per_capita'], kde=True)
plt.title("GDP Per Capita Distribution")
plt.show()

# b. Scatter plot
sns.scatterplot(x='total_cases', y='gdp_per_capita', data=data)
plt.title("Total Cases vs GDP Per Capita")
plt.show()

# c. Pairplot
sns.pairplot(data_groupby)
plt.show()

# d. Bar plot
sns.catplot(data=data_groupby, x='continent', y='total_cases', kind='bar')
plt.title("Total Cases by Continent")
plt.show()
```
